```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as sts
import math
%matplotlib inline
```

В качестве исходного распределения возмем треугольное с параметром c = 0.87

```python
c = 0.87
sample = sts.triang.rvs(c, size = 1000, random_state = 1)
```

Строим нормированную гистограмму по выборке функцию плотности:

```python
x = np.linspace(-0.05, 1.05, 100)
```

```python
plt.hist(sample, bins = 50, normed = True)
plt.plot(x, sts.triang.pdf(x, c))
```

Среднее и дисперсия исходного распределения:

```python
tMean = sts.triang.mean(c)
tVar = sts.triang.var(c)
print("Среднее: ", tMean)
print("Дисперсия: ", tVar)
```

Для выборок размера 5, 10 и 50 строим нормированную гистрограмму распределения выборочных средних и соответсвующую функцию плотности нормального расппределения. Среднее нормального распределения совпадает со средним исходного распределения, а стандартное отклонение вычисляется по формуле math.sqrt(tVar/n) согласно центральной предельной теореме.

```python
n = 5
sampleMeans = list(np.mean(sts.triang.rvs(c, size = n)) for x in range(1000))
plt.hist(sampleMeans, bins = 30, normed = True)
plt.plot(x, sts.norm.pdf(x, loc = tMean, scale = math.sqrt(tVar/n)))
plt.title("n=" + str(n) + " sd=" + str(round(math.sqrt(tVar/n), 3)))
plt.xlabel('$x$')
```

```python
n = 10
sampleMeans = list(np.mean(sts.triang.rvs(c, size = n)) for x in range(1000))
plt.hist(sampleMeans, bins = 30, normed = True)
plt.plot(x, sts.norm.pdf(x, loc = tMean, scale = math.sqrt(tVar/n)))
plt.title("n=" + str(n) + " sd=" + str(round(math.sqrt(tVar/n), 3)))
plt.xlabel('$x$')
```
